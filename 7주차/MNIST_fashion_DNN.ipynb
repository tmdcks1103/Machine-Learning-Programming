```python
# MNIST 패션 데이터 DNN으로 구현

import tensorflow as tf
from tensorflow.keras import models, layers
import numpy as np
import matplotlib.pyplot as plt

# 데이터 로드
fashion_mnist = tf.keras.datasets.fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
# 정규화
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
#DNN 모델 구성
model = models.Sequential([
    # Flatten 층: 28x28 픽셀의 2차원 이미지를 784개의 1차원 배열로 변환.
    layers.Flatten(input_shape=(28, 28)),

    # 첫 번째 Dense 층 (은닉층): 256개의 뉴런을 가지며, relu 활성화 함수 사용.
    layers.Dense(256, activation='relu'),
    # Dropout 층: 학습 시 뉴런의 30%를 비활성화하여 과적합 방지.
    layers.Dropout(0.3),

    # 두 번째 Dense 층 (은닉층)
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),

    # 출력층: 10개의 클래스에 대한 확률을 출력하기 위해 10개의 뉴런과 softmax 활성화 함수를 사용합니다.
    layers.Dense(10, activation='softmax')
])

model.summary()
```

```python
# 컴파일
model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
```

```python
# 모델 학습
history = model.fit(x_train, y_train,
                    epochs=10,
                    batch_size=128,
                    validation_split=0.1, # 훈련 데이터의 10%를 검증용으로 사용
                    verbose=2)
```

```python
# 모델 평가
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"\n✅ 테스트 정확도: {test_acc:.4f}")
```
