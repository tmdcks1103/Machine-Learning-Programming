```python
# MNIST 패션 데이터 CNN으로 구현

import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt

# 데이터 로드
fashion_mnist = tf.keras.datasets.fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
# 정규화
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
# CNN 입력을 위한 채널 차원 추가

x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)

print(f"변환 후 x_train 형태: {x_train.shape}")
print(f"변환 후 x_train 형태: {x_test.shape}")
```

```python
# CNN 모델 구성
model = models.Sequential ([

    # 첫 번째 Conv-Pool 블록: 32개의 3x3 필터로 특징을 추출하고, Max Pooling으로 크기를 줄입니다.
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),

    # 두 번째 Conv-Pool 블록: 64개의 3x3 필터로 더 복잡한 특징을 학습합니다.
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # 분류를 위한 DNN 부분
    # Flatten 층: 다차원 특징 맵을 1차원으로 변환하여 Dense 층에 전달합니다.
    layers.Flatten(),

    # Dense 층 (분류기): 추출된 특징을 바탕으로 이미지를 분류합니다.
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.4), # 과적합 방지
    layers.Dense(10, activation='softmax') # 10개 클래스에 대한 확률 출력
])

model.summary()
```

```python
# 컴파일
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# 모델 학습
history = model.fit(x_train, y_train,
                    epochs=10,
                    batch_size=128,
                    validation_split=0.1,
                    verbose=2)
```

```python
# 모델 평가
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"\n✅ 테스트 정확도 (CNN): {test_acc:.4f}")
```
