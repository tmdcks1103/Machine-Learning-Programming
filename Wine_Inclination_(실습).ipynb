```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/ColabNotebooks/data/winequality-white.csv", sep=';')

df
```

```python
df['quality'].value_counts()
```

```python

#입력과 출력 나누기
X = df.drop(columns=['quality'])
Y = df['quality']

X
```

```python
Y
```

```python
# 결측치 0부터 인코딩

from sklearn.preprocessing import LabelEncoder

# LabelEncoder 객체 생성
encoder = LabelEncoder()

# Y 데이터를 인코딩하여 Y_encoded 변수에 저장
Y_encoded = encoder.fit_transform(Y)

# 변환된 라벨을 Y에 할당하여 업데이트
Y = Y_encoded

# 업데이트된 Y의 상위 5개 값 확인
print("업데이트된 Y 데이터의 상위 5개:\n", Y[:5])

# 업데이트된 Y의 고유한 값과 개수 확인
print("\n업데이트된 Y의 고유한 값들:", pd.Series(Y).unique())
print("\n업데이트된 Y의 값별 개수:\n", pd.Series(Y).value_counts().sort_index())
```

```python
Y
```

```python
#train과 test 나누기
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2, random_state=7 ,stratify=Y)


X
```

```python
Y
```

```python
# 모델 훈련

from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

#DT
dt = tree.DecisionTreeClassifier();
dt.fit(X_train,Y_train)

#RF
rf = RandomForestClassifier();
rf.fit(X_train,Y_train)

#LR
lr = LogisticRegression();
lr.fit(X_train,Y_train)

#KNN
knn = KNeighborsClassifier();
knn.fit(X_train,Y_train)
```

```python
# 정확도 분석 및 기울기, 절편 구하기

from sklearn.metrics import accuracy_score

#DT
Y_pred_dt = dt.predict(X_test)
print("DT 정확도:{}".format(accuracy_score(Y_test,Y_pred_dt)))

#RF
Y_pred_rf = rf.predict(X_test)
print("RF 정확도:{}".format(accuracy_score(Y_test,Y_pred_rf)))

#LR
Y_pred_lr = lr.predict(X_test)
print("LR 정확도:{}".format(accuracy_score(Y_test,Y_pred_lr)))
print("LR 계수:", lr.coef_)
print("LR 절편:", lr.intercept_)

#KNN
Y_pred_knn = knn.predict(X_test)
print("KNN 정확도:{}".format(accuracy_score(Y_test,Y_pred_knn)))
```

```python
# 로지스틱 회귀의 계수(기울기)와 절편 시각화
from sklearn.preprocessing import StandardScaler
from sklearn.inspection import DecisionBoundaryDisplay
import matplotlib.pyplot as plt

# 시각화를 위해 'alcohol'과 'density' 두 특성만 선택합니다.
X_selected = X[['alcohol', 'density']]
Y_selected = Y

# 데이터 스케일링 (로지스틱 회귀 성능 향상 및 시각화를 위해 필요)
scaler = StandardScaler()
X_selected_scaled = scaler.fit_transform(X_selected)

# 다시 훈련/테스트 데이터로 분할합니다.
X_train_sub, X_test_sub, Y_train_sub, Y_test_sub = train_test_split(
    X_selected_scaled, Y_selected, test_size=0.2, random_state=7, stratify=Y_selected
)

# 새로운 로지스틱 회귀 모델을 훈련합니다.
lr_vis = LogisticRegression(max_iter=1000)
lr_vis.fit(X_train_sub, Y_train_sub)

# 기울기(계수)와 절편 출력
print("\n--- 로지스틱 회귀 시각화 모델의 기울기와 절편 ---")
print(f"계수 (기울기): {lr_vis.coef_}")
print(f"절편: {lr_vis.intercept_}")

# 그래프 생성
plt.figure(figsize=(10, 7))

# 결정 경계 시각화
# DecisionBoundaryDisplay는 모델의 예측을 기반으로 결정 경계를 자동으로 그려줍니다.
DecisionBoundaryDisplay.from_estimator(
    lr_vis,
    X_train_sub,
    cmap=plt.cm.coolwarm,
    alpha=0.8,
    ax=plt.gca(),
    xlabel='Alcohol (Standardized)',
    ylabel='Density (Standardized)'
)

# 훈련 데이터 포인트 시각화
plt.scatter(X_train_sub[:, 0], X_train_sub[:, 1], c=Y_train_sub, cmap=plt.cm.coolwarm, edgecolors='k')
plt.title('로지스틱 회귀 모델의 결정 경계 (Alcohol vs. Density)')
plt.show()
```
